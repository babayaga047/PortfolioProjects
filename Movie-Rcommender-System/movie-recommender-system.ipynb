```python
import ast
def convert(x):
    L=[]
    for i in ast.literal_eval(x):
        L.append(i['name'])
    return L

#since we were getting a string as an iterable object, the loop wasn't working, so to convert that string to list we used ast module
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convert5(x):
    L = []
    counter = 0
    for i in ast.literal_eval(x):
        if counter < 5:
            L.append(i['name'])
        counter+=1
    return L 
#from the cast column we only want the name of top 5 actors hence such code
```

```python
movies['cast']=movies['cast'].apply(convert5)
```

```python
movies.head()
```

```python
movies['release_date']=pd.to_datetime(movies['release_date'])
movies['runtime']=movies['runtime'].astype(int)
movies['popularity']=movies['popularity'].astype(int)
```

```python
movies.head()
```

```python
movies.info()
```

```python
def fetch_director(x):
    L = []
    for i in ast.literal_eval(x):
        if i['job']=='Director':
                L.append(i['name'])
                break
    return L 

#taking only director from crew
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())

#converting the overview column which was a string into list in order to concatenate with other new list columns formed
```

```python
movies.head()
```

```python
movies['genres']=movies['genres'].apply(lambda a:[i.replace(" ","")for i in a])
movies['keywords']=movies['keywords'].apply(lambda b:[i.replace(" ","")for i in b])
movies['cast']=movies['cast'].apply(lambda c:[i.replace(" ","")for i in c])
movies['crew']=movies['crew'].apply(lambda d:[i.replace(" ","")for i in d])

#removing spaces betweem two words, so that for eg sam worthington and sam mendes can be treated as different entites in tag
```

```python
movies.rename(columns={"runtime":"runtime (in_mins)"},inplace=True)
```

```python
movies.head()
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
movies.head()
```

```python
new_df=movies[['id','title','tags']]
new_df
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))

#convert the list of tag column into string
```

```python
new_df.head()
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
new_df['tags'][0]
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()

def stem(x):
    y=[]
    for i in x.split():
        y.append(ps.stem(i))
    return " ".join(y)
        
#using stemming technique here, x are the words going into the function one by one, then converting the string of words into a list
#while applying stemming function and then converting the list back into a string

new_df['tags']=new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

```python
#initiating vectorization

from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
vectors[0]

#apllying a feature transformation technique "countvectorizer"
#converting to array since the countvec contains alot of zeroes stored in the form of a sparse matrix
#creating the text to numbers; the shape (4805, 5000) shows 4805 movies with occurence of "top 5000 most similar words from the data" for each movie in the form of numbers each row forming a unique vectror
```

```python
cv.get_feature_names_out()
#np.set_printoptions(threshold=np.inf)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
#similarity inv proportional to distance, hence cosine_distance is calculating angles between two vectors since eucledian distances doesn't work good in higher dimensions
```

```python
similarity=cosine_similarity(vectors)
similarity.shape
#storing the distances in the variable similarity
```

```python
similarity[0]

#tells the distance of 1st movie to each movie
```

```python
list(enumerate(similarity[0]))

#since after sorting the index position of distances will change in each array, so in order to hold the index positions of each value
#showing us in a list of tupples the movie index along with its distances from other movies
#after sorting 1st movie ka 1st movie ke saath, 2nd movie ke saath distance....etc ye pattern we will loose
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[0:6]

#the lamda function is done so to tell the computer that we want to sort on the basis of 2nd number in each tupple which is the distance and not the first which is index of movies
#then we will extract top 5 movies
```

```python
def recommend(movie):
    movie_index=new_df[new_df['title'] == movie].index[0]
    movies_list=sorted(list(enumerate(similarity[movie_index])),reverse=True,key = lambda x: x[1])
    
    for i in movies_list[1:11]:
        print(new_df.iloc[i[0]].title)
        

#when we'll be given a movie, we will be required to first find the index of that movie
#from that index we will enter into the similarity matrix
#in the similarity array for each movie we will first sort each values or you can say distances in that array in descending order i.e. most similar to least similar
#movies_list is storing top 10 distances along with thier id
#then we'll fetch the top 5 movies based on the top 5 distances
```

```python
recommend('Avatar')
```
